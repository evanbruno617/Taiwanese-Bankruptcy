```python
data_df.drop(columns=[' Net Income Flag'], inplace=True)
```

```python
# Define test_df features set.
X = data_df.copy()
X = X.drop(columns=['Bankrupt?'], axis=1)

#X = X[X.columns.tolist()[:-1]]
X.head()
```

```python
# Define the target set.
y = data_df['Bankrupt?'].ravel()
y[:5]
```

```python
# Import our dependencies
import pandas as pd
import matplotlib as plt
import sklearn as skl
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
from imblearn.combine import SMOTEENN
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
columns_to_scale = X.columns.tolist()

columns_to_scale.remove(' Liability-Assets Flag')
#columns_to_scale.remove('Sentiment')
#columns_to_scale.remove('Scores')
#columns_to_scale.remove('Sentiment_stop')
#columns_to_scale.remove('Scores_stop')
```

```python
# Use sklearn to split dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)

ran = RandomOverSampler(random_state=0)
X_train, y_train = ran.fit_resample(X_train, y_train)

# Creating a StandardScaler instance.
X_scaler = StandardScaler()
# Fitting the Standard Scaler with the training data.
X_scaler.fit(X_train[columns_to_scale])
X_train[columns_to_scale] = X_scaler.transform(X_train[columns_to_scale])
X_test[columns_to_scale] = X_scaler.transform(X_test[columns_to_scale])


# Scaling the data.
X_train_scaled = X_scaler.transform(X_train[columns_to_scale])
X_test_scaled = X_scaler.transform(X_test[columns_to_scale])

# Create a random forest classifier.
rf_model = RandomForestClassifier(n_estimators=128, random_state=78) 


# Fitting the model
rf_model = rf_model.fit(X_train, y_train)

# Making predictions using the testing data.
predictions = rf_model.predict(X_test)

# Calculating the confusion matrix.
cm = confusion_matrix(y_test, predictions)
```

```python
cm
```

```python
# Displaying results
print("Confusion Matrix")
display(cm)
print("Classification Report")
print(classification_report(y_test, predictions))
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='lbfgs', random_state=1)
model.fit(X_train, y_train)
```

```python
# Making predictions using the testing data.
predictions = model.predict(X_test)

# Calculating the confusion matrix.
cm = confusion_matrix(y_test, predictions)
cm
```

```python
# Displaying results
print("Confusion Matrix")
display(cm)
print("Classification Report")
print(classification_report(y_test, predictions))
```

```python
# L1 penalty for Lasso
log_lasso = LogisticRegression(penalty='l1', solver='saga', C=1.0, max_iter=1000)  
log_lasso.fit(X_train, y_train)
```

```python
# Making predictions using the testing data.
predictions = log_lasso.predict(X_test)

# Calculating the confusion matrix.
cm = confusion_matrix(y_test, predictions)
cm
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'C': [0.001, 0.01, 0.1, 1, 10, 100]}
grid = GridSearchCV(LogisticRegression(penalty='l1', solver='saga', max_iter=1000), param_grid, cv=5)
grid.fit(X_train, y_train)
print(grid.best_params_)
```
