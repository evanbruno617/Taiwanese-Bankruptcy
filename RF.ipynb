```python
features = pd.DataFrame(zip(rf_model.feature_importances_, X.columns))
test_low = features.sort_values(by=0, ascending=True)
```

```python


results_dict = {'Columns': [], 'Validation_Accuracy': [], 'Recall': []}

for i in test_low[1].tolist():
    X.drop(columns=[i], inplace=True)
    
    if i in columns_to_scale:
        columns_to_scale.remove(i)
    
    # Create scaler instance
    X_scaler = StandardScaler()
    
    # Splitting into Train and Test sets.
    # Splitting into Train and Test sets.
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=48)
    
    X_scaler.fit(X_train[columns_to_scale])
    

    ran = RandomOverSampler(random_state=0)
    X_train, y_train = ran.fit_resample(X_train, y_train)


    # Scaling the data.
    # Scaling the data.
    X_train[columns_to_scale] = X_scaler.transform(X_train[columns_to_scale])
    #X[columns_to_scale] = X_scaler.transform(X[columns_to_scale])


    # Create a random forest classifier.
    rf_model = RandomForestClassifier(n_estimators=500, random_state=78) 


    # Fitting the model
    rf_model = rf_model.fit(X_train, y_train)
    
    X_test[columns_to_scale] = X_scaler.transform(X_test[columns_to_scale])

    # Making predictions using the testing data.
    predictions = rf_model.predict(X_test)
    
    acc = float(classification_report(y_test, predictions).split('\n')[3].split(' ')[18])
    
    recall = float(classification_report(y_test, predictions).split('\n')[3].split(' ')[24])
    
    results_dict['Columns'].append(X.columns.tolist())
    results_dict['Validation_Accuracy'].append(acc)
    results_dict['Recall'].append(recall)
    
    
```

```python
results_df = pd.DataFrame.from_dict(results_dict).sort_values(by='Validation_Accuracy', ascending=False)
results_df
```
